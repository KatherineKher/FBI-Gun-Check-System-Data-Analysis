### Data Cleaning (Replace this with more specific notes!)

#### Gun data:
<ol>
<li>Converting month to datetime and rename it to year-month</li>
<li>Handling Nulls:
<ul> 
<li>Columns with more than half their entries missing will be dropped</li>
<li>Premit recheck nulls were replaced by zero</li>
<li>Remaining columns nulls were replaced by their column's mean</li>
</ul>
    </li>
</ol>

#### Census data:
<ol>
<li>Dropping duplicated rows</li>
<li>Transforming Data so that there would be a column for feature, state and value</li>
<li>Cleaning Fact column
    <ul> 
    <li>Change "percent" to % sign</li>
    <li>Split year if found</li>
    </ul>
</li>
<li>Ensure correct datatypes are reflected</li>
</ol>

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.

# make copy to work on
gun_df = gun_df_original

# For gun_df:
# 1- converting Month to date datatype and rename to year-month
gun_df['month'] = pd.to_datetime(gun_df['month'])
gun_df.rename(columns={'month': 'year-month'}, inplace=True)

# confirm datatype conversion & renaming
gun_df.dtypes
```

```python
# 2- Manage NULLs, first i will drop columns with missing entries > row_count/2 ..i.e. more than half of the entries are missing

# I found help on how to do it here: 
# https://stackoverflow.com/questions/31614804/how-to-delete-a-column-in-pandas-dataframe-based-on-a-condition

# however i modified it a little so that it is applied only from the 5th column onward
gun_df.drop(gun_df.iloc[:, 4:].columns[gun_df.iloc[:, 4:].apply(lambda col: col.isnull().sum() > gun_df.shape[0]/2)], 
            axis=1, inplace=True)
```

```python
# check nulls
gun_df.isnull().sum()
```

```python
# for permit_recheck i will replace nulls with zeros
gun_df['permit_recheck'].fillna(0, inplace=True);

# replace remaining nulls with mean of that column
# found help on how to achieve this from: https://stackoverflow.com/questions/34913590/fillna-in-multiple-columns-in-place-in-python-pandas
gun_df.fillna({'permit': gun_df['permit'].mean(), 
               'handgun': gun_df['handgun'].mean(),
               'long_gun': gun_df['long_gun'].mean(),
               'admin': gun_df['admin'].mean(),
               'prepawn_handgun': gun_df['prepawn_handgun'].mean(),
               'prepawn_long_gun': gun_df['prepawn_long_gun'].mean(),
               'redemption_handgun': gun_df['redemption_handgun'].mean(),
               'redemption_long_gun': gun_df['redemption_long_gun'].mean()}, inplace=True)

# check nulls
gun_df.isnull().sum()
```

```python
# modify totals column to be equal to sum of remaining columns
gun_df['totals'] = gun_df.iloc[:,2:11].sum(axis=1)
```

```python
# make copy of census data to work on
census_df = census_df_original

# for census_df:
# 1- drop duplicates
census_df.drop_duplicates(inplace=True);

# confirm no more duplicates are found
census_df.duplicated().sum()
```

```python
# 2- Transform data so that there would be only 4 coulmns - also take into account that our data is till row 64 only
census_df = census_df.iloc[:64,]
census_df.shape
```

```python
# found a way on how to achieve this from:
# https://stackoverflow.com/questions/28654047/pandas-convert-some-columns-into-rows

census_df = census_df.melt(id_vars=["Fact", "Fact Note"], 
            var_name="State", 
            value_name="Value");
```

```python
# check datetypes of our new transformed df
census_df.State.unique()
```

```python
# Clean census_df.Value of non-numeric values
census_df.Value.replace('\D+','',regex=True, inplace=True)
census_df.dtypes
```

```python
# convert values column to numeric
census_df['Value'] = pd.to_numeric(census_df['Value'])

# confirm change
census_df.dtypes
```

```python
# check for nulls
census_df.isnull().sum()
```

```python
# fill Value with nulls with zeros
census_df['Value'].fillna(0, inplace=True);

# confirm nulls
census_df.isnull().sum()
```

```python
census_df.shape
```

## Exploratory Data Analysis


### Research Question 1 (What is the overall trend of gun purchases?)

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.

# sum totals in gun_df by year to get overall trend of gun registration
gun_df.groupby(['year-month'])['totals'].sum().plot(figsize=(15,10));
mp.suptitle('Sum of guns in all states during entire study');
mp.xlabel('Date');
mp.ylabel('Sum of guns');
```

### Research Question 2  (Which states have had the highest growth in gun registrations?)

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.

# sum totals in gun_df by state to get overall trend of gun registration .. get top 20 countries
# got help on sorting from: 
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html

gun_df.groupby(['state'])['totals'].sum().sort_values(ascending=False)[0:20].plot(kind='bar',figsize=(20,10), color=(0.2, 0.4, 0.6, 0.6));
mp.suptitle('Sum of guns per State', fontsize=18);
mp.xlabel('States', fontsize=14);
mp.ylabel('Sum of guns', fontsize=14)
mp.tick_params(axis='both', labelsize=12);
```

```python
# we now investigate increase of gun registration per state
first_year_data = gun_df[gun_df['year-month']==min(gun_df['year-month'])][['state','totals']]
last_year_data = gun_df[gun_df['year-month']==max(gun_df['year-month'])][['state','totals']]


merged_data = pd.merge(last_year_data, first_year_data,  on='state')
merged_data['difference'] = merged_data['totals_x']-merged_data['totals_y']

merged_data.sort_values(by='difference', ascending=False, inplace=True)

# plot top 20 states which have the highest incearse in gun registration since first data entry till last date
merged_data.groupby(['state'])['difference'].sum().sort_values(ascending=False)[0:20].plot(kind='bar',figsize=(20,10), color=(0.2, 0.4, 0.6, 0.6));
mp.suptitle('Top 20 States with highest difference in gun registration since start till end of this study', fontsize=18);
mp.xlabel('States', fontsize=14);
mp.ylabel('# of gun registration increase', fontsize=14)
mp.tick_params(axis='both', labelsize=12);
```

### Research Question 3  (What census data is most associated with high gun per capita?)

```python
# add a column with idicator of 1 to recognize our top 20 states 
census_df['indicator'] = 'blue'

# check if State is in our merged data, if yes, indicator =1
census_df.loc[census_df['State'].isin(merged_data.state.iloc[0:20].unique()), ['indicator']] = 'red'

# confirm changes
census_df.head()
```

```python
# split census data into groups, like age/race/education/employment factors

# race analysis State per Race

race_filter = ['White alone, percent, July 1, 2016,  (V2016)',
               'Black or African American alone, percent, July 1, 2016,  (V2016)',
               'American Indian and Alaska Native alone, percent, July 1, 2016,  (V2016)',
               'Asian alone, percent, July 1, 2016,  (V2016)',
               'Native Hawaiian and Other Pacific Islander alone, percent, July 1, 2016,  (V2016)',
               'Two or More Races, percent, July 1, 2016,  (V2016)',
               'Hispanic or Latino, percent, July 1, 2016,  (V2016)',
               'White alone, not Hispanic or Latino, percent, July 1, 2016,  (V2016)']

# filter our data by our race filter .. then groupby race & states and sum up values .. finally plot
# got help on muliplot from here: https://stackoverflow.com/questions/45765735/arranging-multiple-for-loop-categorical-plots-with-seaborn

census_df[census_df.Fact.isin(race_filter)].groupby(['State','Fact'])['Value'].sum().sort_values(ascending=False).unstack().plot(kind='bar'
                                                                                                          ,subplots=True
                                                                                                          ,figsize=(15,25) 
                                                                                                          ,layout=(8,1)
                                                                                                          ,legend=True
                                                                                                          ,sharey=True
                                                                                                         ,color='blue')

mp.suptitle('Race % in all States', fontsize=18);
mp.ylabel('Race Percent', fontsize=14)
mp.legend()
mp.tick_params(axis='both', labelsize=12);
```

```python
# split census data into groups, like age/race/education/employment factors
# perform same analysis as Race but for age

# age analysis State per Age
age_filter = ['Persons under 5 years, percent, July 1, 2016,  (V2016)',
              'Persons under 5 years, percent, April 1, 2010',
              'Persons under 18 years, percent, July 1, 2016,  (V2016)',
              'Persons under 18 years, percent, April 1, 2010',
              'Persons 65 years and over, percent,  July 1, 2016,  (V2016)',
              'Persons 65 years and over, percent, April 1, 2010']


# filter our data by our age filter .. then groupby race & states and sum up values .. finally plot
census_df[census_df.Fact.isin(age_filter)].groupby(['State','Fact'])['Value'].sum().unstack().plot(kind='bar'
                                                                                                          ,subplots=True
                                                                                                          ,figsize=(15,25) 
                                                                                                          ,layout=(6,1)
                                                                                                          ,sharey=True
                                                                                                  ,color='blue')
mp.suptitle('Age % in all States', fontsize=18);
mp.ylabel('Age Percent', fontsize=14)
mp.tick_params(axis='both', labelsize=12);
```

```python
# split census data into groups, like age/race/education/employment factors
# perform same analysis as Race & age but for living standards

# view difference between states in 'Per capita income in past 12 months (in 2015 dollars), 2011-2015'
census_df[census_df.Fact == 'Per capita income in past 12 months (in 2015 dollars), 2011-2015'].groupby(['State'])['Value'].sum().sort_values(ascending=False).plot(kind='barh',color=(0.2, 0.4, 0.6, 0.6), figsize=(10,15))
mp.suptitle('Capita income 2011-2015 in Top 20 States', fontsize=18);
mp.ylabel('Capita income', fontsize=14)
mp.legend()
mp.tick_params(axis='both', labelsize=12);
```

```python
# view difference between states in 'Total annual payroll, 2015 ($1,000)'
census_df[census_df.Fact == 'Total annual payroll, 2015 ($1,000)'].groupby(['State'])['Value'].sum().sort_values(ascending=False).plot(kind='barh',color=(0.2, 0.4, 0.6, 0.6), figsize=(10,15))
mp.suptitle('Total annual payroll, 2015 ($1,000) in all States', fontsize=18);
mp.ylabel('Total annual payroll', fontsize=14)
mp.legend()
mp.tick_params(axis='both', labelsize=12);
```

```python
# view difference between states in 'Persons in poverty, percent'
census_df[census_df.Fact == 'Persons in poverty, percent'].groupby(['State'])['Value'].sum().sort_values(ascending=False).plot(kind='barh',color=(0.2, 0.4, 0.6, 0.6), figsize=(10,15))
mp.suptitle('Poverty % in all States', fontsize=18);
mp.ylabel('Poverty income', fontsize=14)
mp.legend()
mp.tick_params(axis='both', labelsize=12);
```

```python
# split census data into groups, like age/race/education/employment factors

# education analysis 

education_filter = ['Language other than English spoken at home, percent of persons age 5 years+, 2011-2015',
               'High school graduate or higher, percent of persons age 25 years+, 2011-2015',
               "Bachelor's degree or higher, percent of persons age 25 years+, 2011-2015"]

# filter our data by our education filter .. then groupby race & states and sum up values .. finally plot
# got help on muliplot from here: https://stackoverflow.com/questions/45765735/arranging-multiple-for-loop-categorical-plots-with-seaborn

census_df[census_df.Fact.isin(education_filter)].groupby(['State','Fact'])['Value'].sum().sort_values(ascending=False).unstack().plot(kind='bar'
                                                                                                          ,subplots=True
                                                                                                          ,figsize=(15,5) 
                                                                                                          ,layout=(3,1)
                                                                                                          ,legend=True
                                                                                                          ,sharey=True
                                                                                                         ,color='blue')

mp.suptitle('Education in all States', fontsize=18);
mp.ylabel('Education %', fontsize=14)
mp.legend()
mp.tick_params(axis='both', labelsize=12);
```
